```python
import pandas as pd
```

```python
print("Loading data...\n\n","-"*60,"\n\n")
df = pd.read_csv("practice.csv",skipinitialspace=True)
print(df)
```

```python
#INFORMATION

print("\nDataset Info:\n")
print(df.info())
print("\nMissing Values: \n")
print(df.isnull().sum())
print("\nDuplicate Rows:", df.duplicated().sum())
```

```python
#REMOVE DUPLICATES

print("\nRemoving duplicate rows...")
df = df.drop_duplicates()
print("\ndone..!")
```

```python
#FILLING MISSING VALUES IN AGE COLOUMN

if df['Age'].isnull().sum() > 0:
  df['Age'] = pd.to_numeric(df['Age'],errors = 'coerce')
  print("\nFilling missing values...")
  average_age = df["Age"].mean()
  df['Age'] = df['Age'].fillna(round(average_age))
```

```python
#FILLING MISSING VALUES IN SALARY COLOUMN

if df['Salary'].isnull().sum() > 0:
  df['Salary'] = pd.to_numeric(df['Salary'], errors = 'coerce')
  median_salary = df['Salary'].median()
  df['Salary'] = df['Salary'].fillna(round(median_salary))
```

```python
#FILLING MISSING VALUES IN EMAIL COLOUMN

if df['Email'].isnull().sum() > 0:
  df['Email'] = df['Email'].fillna("none@example.com")
```

```python
#CLEANING TEXTS COLOUMNS
df['Name'] = df['Name'].str.strip().str.title()
df["Email"] = df["Email"].str.lower().str.strip()
```

```python
#FIXING DATE

df['Join_Date'] = pd.to_datetime(df['Join_Date'], errors='coerce')
df["Join_Date"] = df["Join_Date"].dt.strftime("%Y-%m-%d")
```

```python
#FIX DATATYPES
df["Age"] = df["Age"].astype(int)
df["Salary"] = df["Salary"].astype(int)
```

```python
print("\nCleaned Dataset:\n")
print(df)
```

```python
print("\nSummary Statistics:\n")
print(df.describe())
```

```python
output = "cleaned_data.csv"
df.to_csv(output, index=False)
print("\nCleaned data saved to", output)
```
